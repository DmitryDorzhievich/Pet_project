```python
import math

import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/home/dmitry/reposit/Pet_project/_data.csv'
```

```python
pet = pd.read_csv(path)
```

```python
pet['Адрес']  = pet['Адрес'].str.split(', ')
pet['Адрес'] = pet['Адрес'].apply(lambda x: x[0])
city_counts = pet['Адрес'].value_counts()
plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Город')
plt.ylabel('Количество объявлений')
plt.title('Количество объявлений в городах');
```

## В данном файле находиться объяления о сдаче жилья. Первый график иллюстрирует какая часть относится к городу Москва. В дальнейшем мы используем данные только по городу Москва.

```python
pet = pet[pet['Адрес'] =='Москва']
pet = pet.reset_index(drop=True)
```

```python
lst = []
for item in pet['Цена'].tolist():
    if '$' in item:
        lst.append(95*float(item[:item.find('$')]))
    elif '€' in item:
        lst.append(102*float(item[:item.find('€')]))
    else:
        lst.append(float(item[:item.find('руб')]))
```

```python
pet['price'] = lst
```

```python
fig, ax = plt.subplots()

sns.histplot(data=pet['price'], bins=20) 
plt.xlabel("Цена, млн. руб.")
plt.ylabel("Количество")
plt.title("Гистрограмма по цене");
```

## На данном графике видно, что большая часть объявлений находится в диапазоне до 200 т.р.

```python
pet['Количество комнат'] = pet['Количество комнат'].str.replace(r'\D', '', regex=True)
pet['Количество комнат'] = pd.to_numeric(pet['Количество комнат'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.scatter(pet['Количество комнат'], pet['price'], alpha=0.5)
plt.xlabel('Количество комнат')
plt.ylabel('Цена')
plt.title('Зависимость цены от количества комнат')
```

## Данный график показывает зависимость стоимости аренды от количества комнат. Видно, что с увеличением количества комнат цена растет.

```python
lst2 = []
for j in pet['Площадь, м2'].tolist():
    j1 = j.split('/')
    lst2.append(float(j1[0]))
pet['Общая площадь'] = lst2
fig = plt.figure(figsize=(15, 5), dpi=200)
pet_sorted = pet.sort_values(by='Общая площадь')
plt.plot(pet_sorted['Общая площадь'], pet_sorted['price'])
plt.xlabel('Общая площадь')
plt.ylabel('Цена');Данный
```

## Данный график показывает данные по стоимости аренды от общей площади квартиры.

```python
pet['Метро'] = pet['Метро'].str.replace(r'\D', '', regex=True)
pet['Метро'] = pd.to_numeric(pet['Метро'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.scatter(pet['Метро'], pet['price'], alpha=0.5)
plt.xlabel('Минут до метро')
plt.ylabel('Цена')
plt.title('Зависимость цены от расстояния до метро')
```

## График нужно будет перестроить, избавившись от выбросов.

```python
parking_counts = pet['Парковка'].value_counts()

parking_counts.plot(kind='bar')

plt.xlabel('Значение парковки')
plt.ylabel('Количество')
plt.title('Гистограмма парковки')

for i, count in enumerate(parking_counts):
    plt.annotate(str(count), xy=(i, count), ha='center', va='bottom')
```

## Больше половины парковок наземные, остальные относятся к другим типам. 

```python
parking_counts = pet.groupby('Парковка')['price'].mean()

parking_counts.plot(kind='bar')

plt.xlabel('Значение парковки')
plt.ylabel('Цена')
plt.title('Гистограмма парковки')
```

## Зависимость средней стоимости аренды от типа парковки.

```python
sns.barplot(data=pet, x="price", y="Санузел", errorbar=None)

plt.xlabel("Цена")
plt.ylabel("Санузел")
plt.title("Зависимость цены от типа санузла");
```

## Зависимость стоимости аренды от типа и количества санузла.

```python
pet.isna().sum().plot(kind ='bar')
plt.xlabel('Столбец')
plt.ylabel('Количество нулевых значений')
plt.title('График определния нулевых значений в столбцах');
```

## График отоброжает количество пропущенных данных.
